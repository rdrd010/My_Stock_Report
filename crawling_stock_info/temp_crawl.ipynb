```python
# 데이터 크롤링 모듈
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import re
driver = webdriver.Chrome('webdrivers/chromedriver.exe')
```

```python
#코스피200 목록 추출(2020-11-06)

company200 = []
for i in range(1,21):
    url = 'https://finance.naver.com/sise/entryJongmok.nhn?&page={}'.format(i)
    driver.get(url)
    driver.implicitly_wait(20)
    html = driver.find_element_by_xpath('/html/body/div/table[1]/tbody').get_attribute("innerHTML")  # 기록 table을 str형태로 저장
    soup = BeautifulSoup(html, 'html.parser')  # str 객체를 BeautifulSoup 객체로 변경
    temp = [i.text.strip() for i in soup.findAll("a")]
    temp = temp[7:]
    company200.extend(temp)
for i in company200:
    print(i)
```

```python
cnt = 1
kospi200 = {}
for i in company200:
    url = 'https://www.naver.com/'
    driver.get(url)
    search = driver.find_element_by_xpath('//*[@id="query"]')
    search.send_keys(i)
    search.send_keys(Keys.ENTER)
    driver.implicitly_wait(20)
    html = driver.find_element_by_xpath('//*[@id="_cs_root"]/div[1]/div/h3/a').get_attribute("innerHTML")  # 기록 table을 str형태로 저장
    soup = BeautifulSoup(html, 'html.parser')  # str 객체를 BeautifulSoup 객체로 변경
    name = soup.find("span", class_="stk_nm").text.strip()
    code = soup.find("em", class_="t_nm").text.strip()
    kospi200[code] = name
    print(cnt)
    print(name, ':', code)
    print('--------')
    cnt += 1
```
